```python
from transformers import AutoModelForSeq2SeqLM, Seq2SeqTrainingArguments, Seq2SeqTrainer
from transformers import AutoTokenizer
```

```python
# Import metrics
from evaluate import load

chrf_metric = load('chrf')
bleu_metric = load('sacrebleu')
comet_metric = load('comet')
```

```python
# Sign into HuggingFace
from google.colab import userdata
userdata.get('HuggingFaceKey')

from huggingface_hub import notebook_login

notebook_login('HuggingFaceKey')
```

```python
# Load model
model = AutoModelForSeq2SeqLM.from_pretrained('ryannhally/t5-small-nah-qu-2')
tokenizer = AutoTokenizer.from_pretrained('ryannhally/t5-small-nah-qu-2')
```

```python
with open('val_pairs.txt', 'r') as f:
    val_raw = f.readlines()

val_pairs = []
for l in val_raw:
    pairs = l.replace('\n', '').replace('  ', ' ').split('|')
    spa = pairs[0].strip()
    nah = pairs[1].strip()

    val_pairs.append((spa, nah))
```

```python
def generate(sentence):
    input_ids = tokenizer('translate Spanish to Nahuatl: ' + sentence, return_tensors='pt').input_ids
    outputs = model.generate(input_ids)
    return tokenizer.batch_decode(outputs, skip_special_tokens=True)[0]
```

```python
# Compute Bleu and Chrf from val_pairs.txt

sum_chrf = 0
sum_bleu = 0
n = len(val_pairs)

all_predictions = []
all_references = []
all_sources = []

highest_chrf = 0.0
highest_bleu = 0.0


for spa, nah in val_pairs:
    out = generate(spa)

    #print(out)
    #print(inp)
    #print('reference:', nah)
    #print('prediction:', out)

    predictions = [out]
    references = [[nah]]

    chrf = chrf_metric.compute(predictions=predictions, references=references)
    bleu = bleu_metric.compute(predictions=predictions, references=references)


    #print('ChrF++ :', chrf['score'])
    #print('BLEU: ', bleu['score'])
    #print('\n')

    #print(total['score'])
    #b = sentence_bleu([tar], out)
    #t+= b
    print('Spanish: ', spa,'Prediction: ', out,', chrF: ', chrf['score'], ', sacreBLEU: ', bleu['score'])



    sum_chrf += chrf['score']
    sum_bleu += bleu['score']

    all_predictions.append(out)
    all_references.append(nah)
    all_sources.append(spa)

    if chrf['score'] > highest_chrf:
        highest_chrf = chrf['score']
        best_chrf = out
        best_chrf_reference = references

    if bleu['score'] > highest_bleu:
        highest_bleu = bleu['score']
        best_bleu = out
        best_bleu_reference = references


print('Average ChrF++ score:', sum_chrf/n)
print('Average BLEU score:', sum_bleu/n)

print('Closest translation (ChrF++ ) : \n ', 'score: ', highest_chrf, 'translation: ', best_chrf, 'reference: ', best_chrf_reference)
print('Closest translation (BLEU): \n ', 'score: ', highest_bleu, 'translation: ', best_bleu, 'reference: ', best_bleu_reference)
```

```python
# Compute comet from val_pairs.txt

predictions=all_predictions
references=all_references #all_references
sources=all_sources

comet = comet_metric.compute(sources=sources, predictions=predictions, references=references)

for scores in comet['scores']:
  print(scores)

print('COMET: ', comet['mean_score'])
```
